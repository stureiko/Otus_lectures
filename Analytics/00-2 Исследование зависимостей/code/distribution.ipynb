```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Пример данных (должны быть положительными)
data = np.random.exponential(scale=2, size=1000)

# Преобразование Бокса-Кокса
transformed_data, lambda_value = stats.boxcox(data)

# Параметры нормального распределения для преобразованных данных
mean = np.mean(transformed_data)
std = np.std(transformed_data)

# Теоретическая кривая нормального распределения
x = np.linspace(min(transformed_data), max(transformed_data), 1000)
pdf = stats.norm.pdf(x, mean, std)

# Визуализация исходных данных и преобразованных данных
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, color='skyblue', edgecolor='black')
plt.title('Гистограмма исходных данных')

plt.subplot(1, 2, 2)
plt.hist(transformed_data, bins=30, density=True, color='lightgreen', edgecolor='black')
plt.title(f'Гистограмма преобразованных данных (λ = {lambda_value:.2f})')
# Добавление теоретической кривой на график
plt.plot(x, pdf, color='firebrick', linestyle='--', linewidth=2, label='Теоретическое нормальное распределение')
plt.legend()

plt.show()

# print(f"Оптимальное значение λ: {lambda_value:.2f}")
```
