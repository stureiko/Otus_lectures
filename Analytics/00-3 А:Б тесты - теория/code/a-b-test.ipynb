# Пример одного эксперимента.

Теоретическое распределение выборочных средних и выделенные значения для одного эксперимента

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Размер выборки
n = 10000

# Стандартное отклонение генеральной совокупности
sigma_main = 1

# Генерация первой выборки
mu_1, sigma_1 = 0, .1  # Среднее и стандартное отклонение
data_1 = np.random.normal(mu_1, sigma_1, n)

# Генерация второй выборки
mu_2, sigma_2 = 1, .1  # Среднее и стандартное отклонение
data_2 = np.random.normal(mu_2, sigma_2, n)


# Построение гистограммы данных
plt.figure(figsize=(6, 6))
# count_1, bins_1, ignored_1 = plt.hist(data_1, bins=30, density=True, alpha=0.4, color='skyblue', edgecolor='black')
count_1, bins_1, ignored_1 = plt.hist(data_1, bins=30, density=True, alpha=0.8, color='skyblue')
# Выделение одного из столбцов
highlight_bin = np.random.randint(0, 30)

ignored_1[highlight_bin].set_facecolor('blue')
ignored_1[highlight_bin].set_alpha(1)

count_2, bins_2, ignored_2 = plt.hist(data_2, bins=30, density=True, alpha=0.8, color='bisque')
highlight_bin = np.random.randint(0, 30)
ignored_2[highlight_bin].set_facecolor('darkorange')
ignored_2[highlight_bin].set_alpha(1)


# Построение теоретической плотности вероятности нормального распределения
xmin, xmax = plt.xlim([-1, 2])
x = np.linspace(xmin, xmax, 100)

p_1 = norm.pdf(x, mu_1, sigma_1)
p_2 = norm.pdf(x, mu_2, sigma_2)

plt.plot(x, p_1, color='royalblue', linewidth=1, linestyle='--', label='Эксперимент 1')
plt.plot(x, p_2, color='burlywood', linewidth=1, linestyle='--', label='Эксперимент 2')
# plt.title('Гистограмма значений и теоретическое нормальное распределение')
plt.title('Теоретическое распределение и эксперимент')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid(True)

plt.show()
```

```python
n = 100
xmin = 3
xmax = 30
x = np.linspace(xmin, xmax, n)
y = 1/x**2 + np.random.random(size=n)*(1/x)

plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Experimet')
plt.hlines(y=0.05, xmin=xmin, xmax=xmax, colors='darkred', linestyle='--', label='p-value')
plt.legend()
plt.xlabel('Days')
plt.ylabel('t-statistic')
plt.show()
```
