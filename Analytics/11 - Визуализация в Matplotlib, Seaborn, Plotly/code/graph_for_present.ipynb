```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({'Stadium': ['Arrowhead Stadium,\nKansas City', 'CenturyLink Field,\nSeattle', 'Husky Stadium,\nSeattle', 
                               'Memorial Stadium,\nClemson', 'TurkTelekom Stadium,\nIstanbul'], 
                   'Loudness': [142, 138, 134, 133, 131]})
```

```python
bars = plt.bar(df['Stadium'], df['Loudness'], width=0.6)
plt.ylim([130, 144])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Stadium', fontsize=12)
plt.ylabel('Loudness (dB)', fontsize=12)
plt.title('World\'s Loudnest Stadiums', fontsize=14)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + .3, round(yval, 2), ha='center', va='bottom', fontsize=12)
# plt.savefig('stadiums.png')
plt.show()
```
