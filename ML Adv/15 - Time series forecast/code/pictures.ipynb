```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 100, 1000)
y0 = 0.01 * x
y1 = 0.5 * np.sin(x / 8)
y01=y0 + y1
y2 = 0.1 * np.sin(x)

white_noise = 0.1 * np.random.normal(size=1000)
y02 = y0 + y1 + y2 + white_noise
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
ax.plot(x, y02, color='steelblue')
# plt.savefig('../Presentation/img/clear.png', dpi=300)
plt.show()
```

```python
fig, axs = plt.subplots(5, 1, figsize=(16, 8))
fig.subplots_adjust(hspace=0.7)
axs[0].plot(x, y02, color='steelblue')
axs[0].set_title('Original')
axs[1].plot(x, y0, color='green')
axs[1].set_title('Trend')
axs[2].plot(x, y1, color='red')
axs[2].set_title('Cyclisity')
axs[3].plot(x, y2, color='orange')
axs[3].set_title('Seasonal')
axs[4].plot(x, white_noise, color='teal')
axs[4].set_title('Residuals')
# fig.savefig('../Presentation/img/clear_decomposition.png', dpi=300)
plt.show()
```
