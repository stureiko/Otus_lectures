```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
# Plot the distribution
a = 3
b = 8
x = np.arange (0, 1, 0.01)
y = beta.pdf(x, a, b)
plt.figure(figsize=(11,6))
plt.plot(x, beta.pdf(x, a, b), linewidth=3)
plt.hist(beta.rvs(a, b, size=1000), density=True, bins='auto', histtype='stepfilled', color='green', alpha=0.4)
plt.xlabel(r'$\theta$', fontsize=20);
plt.ylabel(r'$p(\theta|X)$', fontsize=20);
# plt.xticks(fontsize=20)
# plt.yticks(fontsize=20)

# plt.show()
plt.savefig('sampling.jpg')
```
