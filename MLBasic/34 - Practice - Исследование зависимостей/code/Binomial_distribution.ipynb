```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import statsmodels.stats as st
```

# Бернулли

```python
# Случайная величина из распределения Бурнулли (подбрасывание монетки)
# Параметры:
#   p - вероятность положительного исхода
#   n - количество значений в серии
# sp.stats.bernoulli.rvs(p=0.5, size=n) 

# Моделируем броски монетки с использованием scipy.stats.bernoulli
n = 100
rolls = sp.stats.bernoulli.rvs(p=.5, size=n)

# # Считаем количество выпадений каждой грани
counts = np.bincount(rolls, minlength=2)

# Рисуем гистограмму
plt.figure(figsize=(8, 6))
plt.bar(range(0, 2), counts, color='skyblue', edgecolor='black')
plt.xlabel('Грани кубика')
plt.ylabel('Частота выпадения')
plt.title(f'Распределение выпадения монетки при {n} бросках')
plt.xticks(range(0, 2))
plt.ylim([0.9*counts.min(), 1.1*counts.max()])
plt.show()
```

```python
res = []
m = 10    # Число событий в каждом испытании
n = 1000  # Количество испытаний
p = 0.5  # Вероятность успеха
for i in range(n):
    res.append(sp.stats.bernoulli.rvs(p=p, size=m).sum())

counts = np.bincount(res)

plt.figure(figsize=(10, 6))
plt.bar(range(0, len(counts)), counts)
plt.xlabel('Количество успехов')
plt.ylabel('Количество событий')
plt.title(f'Распределение числа выпадения монетки\nчисло испытаний (n) = {n},\
            \nчисло событий в каждом испытании (m) = {m},\nвероятность успеха у одном событии (p) = {p})')
plt.show()
```

# Биномиальное

Биномиальное распределение — это дискретное распределение, которое описывает количество успехов в серии независимых испытаний Бернулли, где каждый эксперимент имеет два возможных исхода: успех (с вероятностью p) или неудача с вероятностью (1−p). 

Это распространённая модель в задачах, где необходимо подсчитать количество успехов за фиксированное количество попыток.

```python
# Параметры
n = 100  # Количество испытаний
m = 10   # Число событий в каждом испытании
p = 0.5  # Вероятность успеха

# Генерация биномиального распределения
x = np.arange(0, m+1)
pmf = sp.stats.binom.pmf(x, m, p)

# Построение графика PMF
plt.figure(figsize=(8, 6))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.xlabel('Количество успехов')
plt.ylabel('Вероятность')
plt.title(f'Биномиальное распределение\nчисло испытаний (n) = {n},\nчисло событий в каждом испытании (m) = {m},\nвероятность успеха у одном событии (p) = {p})')
plt.xticks(range(0, m+1, m//10))
plt.grid(True, axis='y')
plt.show()
```
