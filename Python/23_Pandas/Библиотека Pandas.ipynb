Можно добавлять столбцы в Датафрейм путем добавления нового Series или проведения армифметических операций между существующими столбцами

# **Группировка и агрегация**

Метод группировки данных groupby позволяет сгруппировать необходимые показатели по строке. При этом мы можем задать показатель, который хотим видеть в результатах группировки, а далее - функцию, по которой будет проводится вычисление.

```python
#Описательная статистика по сгруппированному значению параметра Churn
columns_to_show=['Total day minutes', 'Total eve minutes', 'Total night minutes']
df.groupby(['Churn'])[columns_to_show].mean()
```

Метод .agg позволяет объединить несколько применяемых функций для группировки

```python
#Описательная статистика по сгруппированному значению параметра Churn через функцию agg и библиотекой Numpy
columns_to_show=['Total day minutes', 'Total eve minutes', 'Total night minutes']
df.groupby(['Churn'])[columns_to_show].agg([np.mean,np.max,np.std,np.min])
```

Метод crosstab выводит так называемую матрицу сопряженностей (confusion matrix).Метод помогает узнать кол-во или частоту пересечения разных показателей. Обычно используется для показателей с  булевыми значениями 

```python
#Таблица сопряженностей для показателей International plan и Churn, относительные показатели
pd.crosstab(df['Churn'],df['International plan'])
```

Создадим 2 датафрейма для проверки работы метода **merge**

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
df1
```

Используем метод merge для слияния двух фреймов. Обратите внимание, что не указано ни столбца ,по котором нужно проводить слияние, ни метода слияния. По умолчанию, merge использует совпадающие имена столцов и метод слияния 'inner'

image.png

```python
pd.merge(df1, df2)
```

Обратите внимание, что я не указал, по какому столбцу производить соединение. В таком случае merge использует в качестве ключей столбцы с одинаковыми именами.

Внешнее левое соединение

```python
pd.merge(df1, df2, how='left')
```

Внешнее правое соединение

```python
pd.merge(df1, df2,how='right')
```

Внешнее соединение

```python
pd.merge(df1, df2,how='outer')
```

Если имена столбцов соединения отличаются, то следует задать их явным образом.

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

У merge есть параметр suffixes, которые позволяет задать строки, которые будут дописываться в конце одинаковых имен колонок в левом и правом Датафреймах

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

Иногда ключ соединения объектов находятся: один в индексе, другой в столбце Датафрема. Тогда можно воспользоваться параметром left_index, right_index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
left1
```

```python
right1.reset_index()
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

В Pandas также существует метод join. У него немного другая конструкция и другие параметры соединения по умолчанию. Если явно не указывать колонку, по которой соединяем, метод будет соединять по индексу. Также по умолчанию метод будет использовать левое внешнее соединение. Кроме того, метод join умеет соединять только по одинаковым названиям в двух датафреймах. 

```python
left1.join(right1)
```

```python
left1.join(right1,on='key',how='outer')
```

В отличии от join и merge, которые по умолчанию работают со столбцами, concat позволяет выбрать, хотим ли мы выполнять объединение по столбцам или по строкам. Для этого в аргументе функции необходимо прописать axis=0 или axis=1, в первом случае вторая таблица будет присоединена к первой снизу, во втором – справа.

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                  'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```
